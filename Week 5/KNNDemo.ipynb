```python
import numpy as np
import pandas as pd
```

```python
dataFrame = pd.read_excel(datasetPath)
```

```python
print(dataFrame)
```

```python
dataSet = dataFrame.to_numpy()
```

```python
print(dataSet)
```

```python
dataSet[np.where(dataSet=='High')] = 2
dataSet[np.where(dataSet=='Medium')] = 1
dataSet[np.where(dataSet=='Low')] = 0
```

```python
print(dataSet)
```

```python
x=dataSet[0:10,0:4]
print(x)
```

```python
y=dataSet[0:10,4]
print(y)
```

```python
x_test=np.asarray([4,3,1505,5])
```

```python
def eucledianDistance(x1,x2):
    distance = (x1[0]-x2[0])**2+(x1[1]-x2[1])**2+(x1[2]-x2[2])**2+(x1[3]-x2[3])**2
    distance = distance**0.5
    return distance
```

```python
eucledianDistance(x[0],x_test)
```

```python
distances=[]

for i in x:
    d = eucledianDistance(i,x_test)
    distances.append(d)

distances= np.asarray(distances)
print(distances)
```

```python
k = 3
sortedDistances = np.argsort(distances)
print(sortedDistances)
```

```python
sortedDistancs = sortedDistances[0:k]
print(sortedDistancs)
```

```python
poll = dict()
poll[0]=0 # voting poll for low
poll[1]=0 # voting poll for medium
poll[2]=0 # voting poll for high
```

```python
for i in sortedDistancs:
    label = y[i]
    poll[label]=poll[label]+1
```

```python
result = max(poll)

if(result==0):
    print('Low rent')
elif(result==1):
    print('Medium rent')
else:
    print('High rent')
```
