```python
import numpy as np
import copy
```

```python
import copy
class State:
    # queenPositions
    def __init__(self,queenPositions=None):
        self.queenPositions = queenPositions
        if self.queenPositions==None:
            self.queenPositions = np.random.choice(range(8), size=(8), replace=False)

    def __str__(self):
        return f'{self.queenPositions}'

    def __eq__(self, other):
        if(other==None):
            return False
        if(np.array_equal(self.queenPositions,other.queenPositions)):
            return True
        else:
            return False

    def moves(self):
        childrens = set()
        for queenNo in range(0,8):
            for move in range(1,8):
                childState=copy.deepcopy(self)
                childState.queenPositions[queenNo]=(childState.queenPositions[queenNo]+move)%8
                childrens.add(childState)


        return childrens


    def __hash__(self):
        return 1
```

```python
def countAttacks(state):
    attackCount=0
    for queenNo in range(0,8):
        for otherQueen in range(0,8):
            if otherQueen==queenNo :
                continue
            # column check # they are already in different rows
            if(state.queenPositions[queenNo]==state.queenPositions[otherQueen]):
                attackCount+=1
            if( abs(state.queenPositions[queenNo]-state.queenPositions[otherQueen]) == abs(queenNo-otherQueen)):
                attackCount+=1
    return attackCount
```

```python
def hill_climbing():
    current= State() # this is an initial state
    while True:

        neighbours = current.moves()
        #for n in neighbours:
        #  print(countAttacks(n))
        bestNeighbour = min(neighbours,key=countAttacks)
        #print(countAttacks(bestNeighbour))

        if(countAttacks(bestNeighbour)<countAttacks(current)):
            current=bestNeighbour
        else:
            return current


x=hill_climbing()
print(x)
```
