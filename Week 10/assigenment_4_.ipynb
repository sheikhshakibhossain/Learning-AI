```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_excel('logistic_regression_dataset.xlsx')
print(df)
```

```python
dataset = df.to_numpy()

normalized_data = np.zeros_like(dataset)
noOfColumns = dataset.shape[1]

for i in range(noOfColumns):
    min_val = np.min(dataset[:, i])
    max_val = np.max(dataset[:, i])
    normalized_data[:, i] = (dataset[:, i] - min_val) / (max_val - min_val)

print(normalized_data)
```

```python
X = normalized_data[:, 0:3]
Y = normalized_data[:, 3].reshape(-1, 1)  # Reshape Y to be (n, 1)

print(X)
print(Y)
```

```python
def sigmoid(z):
    return 1 / (1 + tf.exp(-z))

def lossFunction(yPred, yReal):
    return -tf.reduce_mean(yReal * tf.math.log(yPred) + (1 - yReal) * tf.math.log(1 - yPred))
```

```python
X_tensor = tf.Variable(X, dtype=tf.float64)
Y_tensor = tf.Variable(Y, dtype=tf.float64)

# initial weights and bias
W_tensor = tf.Variable(np.random.randn(3, 1), dtype=tf.float64)
b_tensor = tf.Variable(np.random.randn(1, 1), dtype=tf.float64)

learning_rate = 0.001
epochs = 10000
```

```python
# solution of ques 1
for i in range(epochs):
    with tf.GradientTape(persistent=True) as tape:
        Z = tf.matmul(X_tensor, W_tensor) + b_tensor
        YPred = sigmoid(Z)
        loss = lossFunction(YPred, Y_tensor)

    dlossBydW = tape.gradient(loss, W_tensor)
    dlossBydb = tape.gradient(loss, b_tensor)

    W_tensor.assign_sub(learning_rate * dlossBydW)
    b_tensor.assign_sub(learning_rate * dlossBydb)
    print(f'Epoch {i + 1}, Loss: {loss.numpy()}')

print(W_tensor.numpy())
print(b_tensor.numpy())
```

```python
X_test = X[0:200, ]
Y_test = Y[0:200, ]

X_test_tensor = tf.Variable(X_test, dtype=tf.float64)
Y_test_tensor = tf.Variable(Y_test, dtype=tf.float64)
```

```python
Z_test = tf.matmul(X_test_tensor, W_tensor) + b_tensor
YPred_test = sigmoid(Z_test)
YPred_binary = (YPred_test.numpy() >= 0.5).astype(int)
```

```python
# solution of ques 2
accuracy = np.mean(YPred_binary == Y_test_tensor.numpy())

true_positive = np.sum((YPred_binary == 1) & (Y_test_tensor.numpy() == 1))
predicted_positive = np.sum(YPred_binary == 1)
precision = true_positive / predicted_positive if predicted_positive > 0 else 0

actual_positive = np.sum(Y_test_tensor.numpy() == 1)
recall = true_positive / actual_positive if actual_positive > 0 else 0

f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f'accuracy: {accuracy}')
print(f'precision: {precision}')
print(f'recall: {recall}')
print(f'f1 score: {f1_score}')
```
