```python
import pandas as pd
import numpy as np
```

```python
DATASET_PATH = 'assignment3Dataset.xlsx'
```

```python
df = pd.read_excel(DATASET_PATH)
print(df)
```

```python
dataset = df.to_numpy()
n = dataset.shape[0]
x1 = dataset[ : , 0]
x2 = dataset[ : , 1]
x3 = dataset[ : , 2]
y  = dataset[ : , 3]

print(n)
print(x1)
print(x2)
print(x3)
print(y)
```

```python
P = np.array([
    [n, np.sum(x1), np.sum(x2), np.sum(x3)],
    [np.sum(x1), np.sum(x1**2), np.sum(x1*x2), np.sum(x1*x3)],
    [np.sum(x2), np.sum(x1*x2), np.sum(x2**2), np.sum(x2*x3)],
    [np.sum(x3), np.sum(x1*x3), np.sum(x2*x3), np.sum(x3**2)]
])

Q = np.array([
    [np.sum(y)],
    [np.sum(x1*y)],
    [np.sum(x2*y)],
    [np.sum(x3*y)]
])
```

```python
# if P * A = Q then X = ?
A = np.linalg.solve(P, Q)
print(A)
```

```python
a0 = A[0]
a1 = A[1]
a2 = A[2]
a3 = A[3]

print('coefficients are: ')
print(f'a0={a0}')
print(f'a1={a1}')
print(f'a2={a2}')
print(f'a3={a3}')
```
