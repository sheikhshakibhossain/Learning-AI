```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_file = 'House_Rent_Dataset.csv'
```

```python
data_frame = pd.read_csv(data_file)
data_frame
```

```python
data_array = np.array(data_frame)
data_array = data_array.astype(float)
data_array
```

```python
# x_axis = independent variables
# y_axis = dependent variables
x_axis = data_array[:, 0] # all sq. feet data
y_axis = data_array[:, 1] # all house rent data
plt.scatter(x=x_axis, y=y_axis, color='BLUE', label='Data Points')

# adding lebels and title
plt.xlabel('Sq. feet')
plt.ylabel('House Rent')
plt.title('House rent vs Sq. feet')
plt.legend()

# show the plot
plt.show()
```

```python
# linear regressiion - finding the best startght line
# y = a0 + a1 * x
# a1 = ( n*sum(xi*yi) - sum(xi)*sum(yi) ) / ( n*sum(xi^2) - sum(yi)^2 )

n = data_array.shape[0]
# print(n)

xi_times_yi = np.multiply(x_axis, y_axis)
sum_of_xi_times_yi = np.sum(xi_times_yi)

sum_of_xi = np.sum(x_axis)
sum_of_yi = np.sum(y_axis)

sum_of_xi_square = np.sum(np.multiply(x_axis, x_axis))
square_of_sum_of_xi = np.multiply(sum_of_xi, sum_of_xi)

a1 = ( (n*sum_of_xi_times_yi) - (sum_of_xi*sum_of_yi) ) / ( (n*sum_of_xi_square) - square_of_sum_of_xi )
print(a1)
```

```python
# a0 = ( sum(yi) - a1*sum(xi) ) / n
a0 = ( sum_of_yi - (a1 * sum_of_xi) ) / n
print(a0)
```

```python
# pedicted house rent based on square feet
# y = mx + c
y =  a1*x_axis + a0
print(y)
regression_line = y
```

```python
# scatter graph
plt.scatter(x=x_axis, y=y_axis, color='BLUE', label='Data Points')

# linear regresion line with x_axis vs regression_line
plt.plot(x_axis, regression_line, color='RED', label=f'Regression line Y = {a0} + {a1}*x')

# adding lebels and title 
plt.xlabel('Sq. feet')
plt.ylabel('House Rent')
plt.title('House rent vs Sq. feet')
plt.legend()


# show the plot
plt.show()
```
