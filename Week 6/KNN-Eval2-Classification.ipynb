```python
dataset = dataframe.to_numpy()
```

```python
print(dataset)
```

```python
y = dataset[0:15,6]
print(y)
```

```python
x= dataset[0:15,1:5]
print(x)
```

```python
def eucledianDistance(x1,x2):
    distance = (x1[0]-x2[0])**2+(x1[1]-x2[1])**2+(x1[2]-x2[2])**2+(x1[3]-x2[3])**2
    distance = distance ** 0.5
    return distance
    # x1 is a array with 4 values [study hours , attendance, assignment, research papers]
```

```python
x_test = [2, 20, 1, 1]
```

```python
distances = []

for i in x:
    d=eucledianDistance(i,x_test)
    distances.append(d)

print(distances)
```

```python
minDistanceIndices = np.argsort(distances)
print(minDistanceIndices)
```

```python
k = 3
minDistanceIndices=minDistanceIndices[0:k]
print(minDistanceIndices)
```

```python
poll = dict()
poll['High']=0
poll['Medium']=0
poll['Low']=0



for i in minDistanceIndices:
    label = y[i] # high, low, medium
    poll[label]=poll[label]+1
```

```python
if(poll['High']>poll['Medium'] and poll['High']>poll['Low']):
    print('high')
elif(poll['Medium']>poll['High'] and poll['Medium']>poll['Low']):
    print('Medium')
else:
    print('Low')
```

```python
def applyKNN(x_test):
    distances = []

    for i in x:
        d=eucledianDistance(i,x_test)
        distances.append(d)
    minDistanceIndices = np.argsort(distances)
    poll = dict()
    poll['High']=0
    poll['Medium']=0
    poll['Low']=0

    # voting
    for i in minDistanceIndices:
      label = y[i] # high, low, medium
      poll[label]=poll[label]+1

    if(poll['High']>poll['Medium'] and poll['High']>poll['Low']):
      return 'High'
    elif(poll['Medium']>poll['High'] and poll['Medium']>poll['Low']):
      return 'Medium'
    else:
      return 'Low'
```

```python
x_test = [6, 90, 5, 4]
result = applyKNN(x_test)
print(result)
```

```python
dataset2 = '/content/drive/MyDrive/Class Contents/5. Summer 2024/AI Lab/Class Materials For The Students/Week 6/dataset2.xlsx'
```

```python
testDataFrame = pd.read_excel(dataset2)
print(testDataFrame)
```

```python
testDataset = testDataFrame.to_numpy()
print(testDataset)
```

```python
x_test = testDataset[:,1:5]
print(x_test)
```

```python
y_test = testDataset[:,6]
print(y_test)
```

```python
predictions = []
for i in x_test:
    prediction = applyKNN(i)
    predictions.append(prediction)

print(predictions)
```

```python
print('Predictions:')
print(predictions)
print('Actual Values:')
print(y_test)
```

```python
correctPredictions = 0

for i in range(4):
    if(predictions[i]==y_test[i]):
        correctPredictions=correctPredictions+1
    else:
        correctPredictions=correctPredictions+0
```

```python
print(correctPredictions)
```

```python
accuracy = correctPredictions/4*100
print(accuracy)
```
