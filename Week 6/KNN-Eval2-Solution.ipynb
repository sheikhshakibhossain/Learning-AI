```python
import numpy as np
import pandas as pd
```

```python
dataset1Link = '/content/drive/MyDrive/Class Contents/5. Summer 2024/AI Lab/Class Materials For The Students/Week 6/dataset1.xlsx'
```

```python
dataframe = pd.read_excel(dataset1Link)
print(dataframe)
```

```python
dataset = dataframe.to_numpy()
```

```python
print(dataset)
```

```python
y = dataset[0:15,5]
print(y)
```

```python
x= dataset[0:15,1:5]
print(x)
```

```python
def eucledianDistance(x1,x2):
    distance = (x1[0]-x2[0])**2+(x1[1]-x2[1])**2+(x1[2]-x2[2])**2+(x1[3]-x2[3])**2
    distance = distance ** 0.5
    return distance
    # x1 is a array with 4 values [study hours , attendance, assignment, research papers]
```

```python
x_test = [6, 90, 5, 4]
```

```python
distances = []

for i in x:
    d=eucledianDistance(i,x_test)
    distances.append(d)

print(distances)
```

```python
minDistanceIndices = np.argsort(distances)
print(minDistanceIndices)
```

```python
k = 3
minDistanceIndices=minDistanceIndices[0:k]
print(minDistanceIndices)
```

```python
# First Method To Find The Average
sum = 0
for i in minDistanceIndices:
    print(y[i])
    sum+=y[i]

print(sum)
average  = sum/k
print(average)
```

```python
# Second Method To Find The Average
closestMarks=y[minDistanceIndices]
averageMarks = np.average(closestMarks)
print(averageMarks)
```

```python
def applyKNN(x_test):
    distances = []
    for i in x:
        d=eucledianDistance(i,x_test)
        distances.append(d)
    #print(distances)
    minDistanceIndices = np.argsort(distances)
    minDistanceIndices=minDistanceIndices[0:k]
    #print(minDistanceIndices)
    closestMarks=y[minDistanceIndices]
    averageMarks = np.average(closestMarks)
    return averageMarks
```

```python
x_test = [6, 90, 5, 4]
result = applyKNN(x_test)
print(result)
```

```python
dataset2 = '/content/drive/MyDrive/Class Contents/5. Summer 2024/AI Lab/Class Materials For The Students/Week 6/dataset2.xlsx'
```

```python
testDataFrame = pd.read_excel(dataset2)
print(testDataFrame)
```
