```python
import numpy as np
import pandas as pd
```

```python
train_dataset = 'Bus_Fare_Dataset.csv'
test_dataset = 'Bus_Fare_Dataset.csv'
```

```python
train_df = pd.read_csv(train_dataset)
test_df = pd.read_csv(test_dataset)
```

```python
print(train_df)
print('#####################################################################')
print(test_df)
```

```python
train_df = train_df.to_numpy()
test_df = test_df.to_numpy()
record = np.array(test_df)

train_df[np.where(train_df=='Yes')] = 1
train_df[np.where(train_df=='No')] = 0

test_df[np.where(test_df=='Yes')] = 1
test_df[np.where(test_df=='No')] = 0

train_df[np.where(train_df=='Available')] = 1
train_df[np.where(train_df=='Not Available')] = 0

test_df[np.where(test_df=='Available')] = 1
test_df[np.where(test_df=='Not Available')] = 0

train_df[np.where(train_df=='Budget')] = 0
train_df[np.where(train_df=='Affordable')] = 1
train_df[np.where(train_df=='Premium')] = 2
train_df[np.where(train_df=='Luxury')] = 3

test_df[np.where(test_df=='Budget')] = 0
test_df[np.where(test_df=='Affordable')] = 1
test_df[np.where(test_df=='Premium')] = 2
test_df[np.where(test_df=='Luxury')] = 3

print(train_df)
print('#####################################################################')
print(test_df)
```

```python
train_x=train_df[:, 0:4]
print(train_x)
```

```python
train_y=train_df[:,4]
print(train_y)
```

```python
def eucledianDistance(x1,x2):
    distance = (x1[0]-x2[0])**2+(x1[1]-x2[1])**2+(x1[2]-x2[2])**2+(x1[3]-x2[3])**2
    distance = distance**0.5
    return distance
```

```python
print(eucledianDistance(train_x[0], train_x[0]))
print(eucledianDistance(train_x[0], train_x[7]))
```

```python
test_x = test_df[:, 0:4]
test_y = test_df[:, 4] # lebel 
```

```python
correct_predictions = []

for i in range(len(test_x)):

    x_test = test_x[i]
    y_test = test_y[i]

    distances=[]

    for i in train_x:
        d = eucledianDistance(i,x_test)
        distances.append(d)

    distances= np.asarray(distances)
    # print(distances)

    k = 3
    sortedDistances = np.argsort(distances)
    # print(sortedDistances)

    sortedDistancs = sortedDistances[0:k]
    # print(sortedDistancs)

    poll = dict()
    poll[0]=0 # voting poll for Budget
    poll[1]=0 # voting poll for Affordable
    poll[2]=0 # voting poll for Premium
    poll[3]=0 # voting poll for Luxury

    for i in sortedDistancs:
        label = train_y[i]
        poll[label]=poll[label]+1

    result = max(poll)

    # if(result==0):
    #     print('Budget')
    # elif(result==1):
    #     print('Affordable')
    # elif(result==2):
    #     print('Premium')
    # else:
    #     print('Luxury')

    # now i have to check if the prediction was right
    # print(result)
    # print(y_test)
    # print(record[0, :])
    if result == y_test:
        # print('True')
        correct_predictions.append(record[i, :])
   

```

```python
print(f'Total correct predictions: {len(correct_predictions)}/{len(test_x)}\n')
print('Correct Prediction rows: ')
for i in correct_predictions:
    print(i)
```
