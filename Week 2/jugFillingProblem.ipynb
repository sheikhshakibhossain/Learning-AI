```python
import copy

class State:
    # variables
    # cL,cR, qL,qR
    # parent
    def __init__(self, qL,qR,cL=3,cR=4):
        self.qL=qL
        self.qR=qR
        self.cL=cL
        self.cR=cR
        self.parent= None

    def __str__(self):
        return f'({self.qL},{self.qR})'

    def emptyLeft(self):
        childState = copy.deepcopy(self) # a copy of the parent is created
        childState.qL=0
        return childState

    def emptyRight(self):
        childState = copy.deepcopy(self) # a copy of the parent is created
        childState.qR=0
        return childState

    def fillLeft(self):
        childState = copy.deepcopy(self) # a copy of the parent is created
        childState.qL=childState.cL
        return childState

    def fillRight(self):
        childState = copy.deepcopy(self) # a copy of the parent is created
        childState.qR=childState.cR
        return childState

    def transferLeftToRight(self):
        childState = copy.deepcopy(self) # a copy of the parent is created

        childState.qR=childState.qR+childState.qL
        childState.qL=0

        if(childState.qR>childState.cR):
            extra = childState.qR - childState.cR
            childState.qR=childState.qR-extra
            childState.qL=extra

        return childState

    def transferRightToLeft(self):
        childState = copy.deepcopy(self) # a copy of the parent is created

        childState.qL=childState.qL+childState.qR
        childState.qR=0

        if(childState.qL>childState.cL):
            extra = childState.qL - childState.cL
            childState.qL=childState.qL-extra
            childState.qR=extra

        return childState

    def __eq__(self,otherState):
        if (otherState == None):
            return False

        if(self.qL==otherState.qL and self.qR==otherState.qR):
            return True
        else:
            return False
    def __hash__(self):
        return self.qL+self.qR
```

```python
state = State (2,2)
child1= state.emptyLeft()
child2= state.emptyRight()
child3= state.fillLeft()
child4= state.fillRight()
child5= state.transferLeftToRight()
child6= state.transferRightToLeft()
```

```python
print(state)
print(child1)
print(child2)
print(child3)
print(child4)
print(child5)
print(child6)
```

```python
def goalTest(state):
    if state == None:
        return False

    if(state.qL==2 or state.qR==2):
        return True
    else:
        return False
```

```python
print(goalTest(state))
print(goalTest(child1))
print(goalTest(child2))
print(goalTest(child3))
print(goalTest(child4))
print(goalTest(child5))
print(goalTest(child6))
```

```python
def bfs (initialNode):
    if goalTest(initialNode):
        return initialNode

    frontier = list() # queue
    explored = set()

    frontier.append(initialNode)

    while True:
        if (len(frontier)==0):
            return None

        node = frontier.pop(0)
        explored.add(node)

        actions = [node.emptyLeft,node.emptyRight,node.fillLeft,node.fillRight,node.transferLeftToRight,node.transferRightToLeft]

        for action in actions:
            child = action()
            child.parent=node
            if (child not in frontier) and (child not in explored):
                if(goalTest(child)):
                    return child
                frontier.append(child)
```

```python
result = bfs(State(0,0))
print(result)
```

```python
def printSequence(result):
    if(result==None):
        return

    printSequence(result.parent)
    print(result)
```
